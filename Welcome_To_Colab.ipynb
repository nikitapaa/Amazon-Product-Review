```
#Load commands
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
Product = pd.read_csv('/content/amazon.csv')      # Load datasets
Product.head(10)
```

```
Product.shape   # Display number of rows and columns
```

```
Product.describe()   # Display summary statistics
```

```
Product.columns   # Display column names
```

```
Product.isnull().sum()   # Check for missing values
```

```
Product.info(10)   # Display basic info and check for missing values
```

```
Product.fillna(method='ffill', inplace=True)   # Fill missing values or drop rows/columns as necessary

```

```
#changing datatypes of the discounted price
Product['discounted_price']=Product['discounted_price'].str.replace(',','')
Product['discounted_price']=Product['discounted_price'].str.replace('₹','')
Product['discounted_price']=Product['discounted_price'].astype(float)

```

```
#Changing datatype of actual price
Product['actual_price']=Product['actual_price'].str.replace(',','')
Product['actual_price']=Product['actual_price'].str.replace('₹','')
Product['actual_price']=Product['actual_price'].astype(float)
```

```
#Changing datatype of discount percentage
Product['discount_percentage']=Product['discount_percentage'].str.replace('%','')
Product['discount_percentage']=Product['discount_percentage'].astype(float)
```

```
#data visualisation
```

```
#Scatterplot between product and rating
plt.figure(figsize=(10,6))
sns.scatterplot(x='discounted_price',y='actual_price',data=Product,color='blue',edgecolor='linen',alpha=0.5)
tick_array=np.arange(0,1000,2000)
plt.xticks(tick_array)
plt.yticks(tick_array)
plt.title('Scatterplot of the discount price and actual price')
plt.xlabel('discount price')
plt.ylabel('actual price')
plt.show()
```


There is decrease in actual price after increase in discount price of the products.

```
#scatterplot of discount_percentage and rating
plt.figure(figsize=(10,6))
sns.scatterplot(x='discount_percentage',y='rating',data=Product,color='blue',edgecolor='linen',alpha=0.5)
tick_array=np.arange(0,100,10)
plt.xticks(tick_array)
plt.yticks(tick_array)
plt.title('Scatterplot of the discount price and rating')
plt.xlabel('discount percentage')
plt.ylabel('rating')
plt.show()
```

Higher the discount percentage higher the purchase number which gives higher ratings of the products. Range between 20-60% discount is showing the higest number of ratings.Increase in discount percentage has resulted in increase of product rating.

```
#Count plot for rating of products.
plt.figure(figsize=(10,6))
sns.countplot(x='rating',data=Product,color='blue',edgecolor='linen',alpha=0.7)
plt.title('Countplot of the Category')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Around 240 number of products has a rating of  4.1.

```
#scatterplot of actual price and rating
plt.figure(figsize=(10,5))
sns.scatterplot(x='actual_price',y='rating',data=Product,color='blue')
tick_array=np.arange(0,1000,2000)
plt.xticks(tick_array)

plt.title('Scatterplot of the actual price and rating')
plt.xlabel('Actual price')
plt.ylabel('Rating')
plt.show()
```

```
The lowest price has the highest rating.Hence, rating decreases as the price increases.
```
